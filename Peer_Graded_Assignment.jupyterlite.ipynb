```python
def create_age_group(age):
    if age >= 70:
        return '>= 70'
    elif age <= 35:
        return '<= 35'
    return '> 35 & < 70'
boston_df['Age group'] = boston_df['AGE'].apply(create_age_group)
sns.boxplot(x=boston_df['Age group'], y= boston_df['MEDV'])
plt.title('MEDV vs AGE variables')
plt.show()
```
